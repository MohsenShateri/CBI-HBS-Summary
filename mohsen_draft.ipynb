```python
import os
import re


col_map = {
    'جمع کل هزینه ناخالص': 'Total_expense',
    'خوراکیها،آشامیدنیها و دخانیات': 'Food_beverages_tobacco',
    'خوراکی ها و آشامیدنی ها': 'Food_beverages',
    'دخانیات': 'Tobacco',
    'پوشاک و کفش': 'Clothing_footwear',
    'مسکن ، آب ، برق و سوخت': 'Housing_utilities_fuel',
    'لوازم، اثاث وخدمات مورد استفاده درخانه': 'Furniture_household_services',
    'درمان و بهداشت': 'Health',
    'حمل و نقل و ارتباطات': 'Transport_communication',
    'حمل و نقل': 'Transport',
    'ارتباطات': 'Communication',
    'تفریح،سرگرمی،تحصیل وخدمات فرهنگی': 'Recreation_education_culture',
    'تفریح و سرگرمی': 'Recreation_entertainment',
    'تحصیل و خدمات فرهنگی': 'Education_cultural_services',
    'کالاها و خدمات متفرقه': 'Misc_goods_services'
}

desired_column_number = 1


def find_all_table_number(years, desired_column_number):
    dfs = {} 
    for i in years:
        for j in os.listdir(f"Data/CSV_Files/{i}"):
            num = int(re.search(r'^جدول_شماره_(\d+)', j).group(1))   
            if num == desired_column_number:
                file_name = pd.read_csv(f"Data/CSV_Files/{i}/{j}")
                dfs[i] = file_name
    return dfs
                
def transpose(df):
    df = df.dropna(how='all')

    first_col = df.iloc[:, 0].astype(str).str.strip()
    # condition: 4 digits, starting with 13 or 14
    ok = first_col.str.fullmatch(r'(13|14)\d{2}').all()
    # first_col = df.iloc[:, 0].astype(str).str.strip()
    # ok = first_col.str.fullmatch(r'(13|14)\d{2}').all()
    
    
    if ("سال" in df.columns) and pd.to_numeric(df["سال"], errors="coerce").notna().all():
        print(2)
        df.rename(columns={"سال": 'year'}, inplace=True)
        cols = ["year"] + [c for c in df.columns if c != "year"]
        df = df[cols]
        return df

    elif not ok:
        print(1)
        df = df.set_index(df.columns[0]).T
        df.index = df.index.astype(str).str.extract(r'(\d+)', expand=False).astype('Int64')
        df = df.reset_index().rename(columns={"index": 'year'})
        df = df.dropna(axis=1, how='all')
        return df

    else:
        print(3)
        df.rename(columns={df.columns[0]: 'year'}, inplace=True)
        df = df.reset_index().drop(columns='index')
        return df


def remove_number_prefix(df: pd.DataFrame) -> pd.DataFrame:
    """
    Remove leading 'number-' patterns like '2-' or '12-' from all column names.
    Returns a new DataFrame (original is unchanged).
    """
    new_df = df.copy()
    new_df.columns = new_df.columns.str.replace(r'^\s*\d+\s*-\s*', '', regex=True)
    return new_df


def get_tabel_year(year,desired_column_number):
    all_table = find_all_table_number([1380, 1390, 1400, 1402],desired_column_number)
    df = all_table[year]
    df = transpose(df)
    df = remove_number_prefix(df)
    df = df.rename(columns=col_map)
    return df

```
