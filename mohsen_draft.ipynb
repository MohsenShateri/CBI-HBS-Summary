```python
INVISIBLE_CHARS = [
        chr(8203),
        chr(173),
        chr(8207),
        chr(8235),
        chr(8236),
        chr(8234),
        chr(65279)
    ]


UNWANTED_SYMBOLS = [
        "\n",
        "\r",
        "\t",
        "…",
        "ـ",
        "_",
        "•",
        "\\*",
        "`",
        "\"",
        "\'",
        "«",
        "»",
        ",",
        ";",
        ":",
    ]


def replace_arabic_characters(s: pd.Series) -> pd.Series:
    """
    Replace Arabic characters with their Farsi equivalents in the Series.
    """
    # mapping for str.translate: keys are ordinals, values are replacement chars
    character_mapping = {
        ord(chr(1610)): chr(1740),  # ي -> ی
        ord(chr(1574)): chr(1740),  # ئ -> ی
        ord(chr(1609)): chr(1740),  # ى -> ی
        ord(chr(1571)): chr(1575),  # أ -> ا
        ord(chr(1573)): chr(1575),  # إ -> ا
        ord(chr(1572)): chr(1608),  # ؤ -> و
        ord(chr(1603)): chr(1705),  # ك -> ک
        ord(chr(1728)): chr(1607),  # ۀ -> ه
        ord(chr(1577)): chr(1607),  # ة -> ه
    }

    return s.astype("string").str.translate(character_mapping)


def sanitize_farsi_text(s: pd.Series) -> pd.Series:
    """
    Clean Farsi text by replacing Arabic characters, removing invisible and unwanted characters,
    normalizing spaces, and stripping leading/trailing spaces.
    """

    s = s.astype("string")

    # 1) normalize Arabic -> Farsi
    s = replace_arabic_characters(s)

    # 2) Replace Zero Width Non-Joiner ('\u200c') with a space
    s = s.str.replace("\u200c", " ", regex=False)

    # 3) Remove other invisible and unwanted characters
    pattern = "[" + "".join(INVISIBLE_CHARS + UNWANTED_SYMBOLS) + "]"
    s = s.str.replace(pattern, "", regex=True)

    # 4) Normalize spaces: collapse multiple spaces to a single space
    s = s.str.replace(r"\s+", " ", regex=True)

    # 5) Strip leading / trailing spaces
    s = s.str.strip()

    # 6) Remove a trailing ".0"
    s = s.str.replace(r"\.0$", "", regex=True)

    return s
```
